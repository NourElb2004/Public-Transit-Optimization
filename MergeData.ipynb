```python
import pandas as pd
existing = pd.read_csv('roads_existing.csv')
# Add missing columns 
existing['Cost'] = None  
existing['IsPotential'] = 0  
potential = pd.read_csv('roads_potential.csv')
# Add missing columns 
potential['Condition'] = None  
potential['IsPotential'] = 1  
existing = existing[['FromID', 'ToID', 'Distance', 'Capacity', 'Condition', 'Cost', 'IsPotential']]
potential = potential[['FromID', 'ToID', 'Distance', 'Capacity', 'Condition', 'Cost', 'IsPotential']]
merged_roads = pd.concat([existing, potential], ignore_index=True)
merged_roads.to_csv('roads.csv', index=False)
print("Merged CSV created as 'roads.csv'")
```

```python
import pandas as pd
locations_df = pd.read_csv("nodes.csv")
facilities_df = pd.read_csv("facilities.csv")
nodes_df = pd.concat([locations_df, facilities_df], ignore_index=True)
nodes_df.to_csv("locations.csv", index=False)
print("✅ Merged loactions saved as locations.csv")
critical_types = [' Medical', ' Government']
critical_nodes = set(nodes_df[nodes_df[' Type'].isin(critical_types)]['ID'].astype(str))
print("✅ Critical Nodes:", critical_nodes)
```
